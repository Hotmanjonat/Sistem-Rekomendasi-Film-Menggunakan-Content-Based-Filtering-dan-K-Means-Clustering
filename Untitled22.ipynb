```python
from google.colab import files
import pandas as pd

# Upload combined_dataset.csv
uploaded = files.upload()

# Load dataset
data = pd.read_csv('combined_dataset.csv')
data.head()
```

```python
# Cek dimensi dataset
print("Dataset Shape:", data.shape)

# Cek kolom
print("Columns:", data.columns.tolist())

# Info dataset
data.info()

# Cek missing values
print("\nMissing Values:\n", data.isnull().sum())
```

```python
# Deskripsi statistik
data.describe()

# Lihat 10 genre paling populer
data['genres'] = data['genres'].str.split('|')
all_genres = pd.Series([genre for sublist in data['genres'] for genre in sublist])
top_genres = all_genres.value_counts().head(10)

print("Top 10 Genres:\n", top_genres)
```

```python
# Split genre hanya untuk baris yang tidak kosong
data['genres'] = data['genres'].dropna().apply(lambda x: x.split('|') if isinstance(x, str) else [])

# Flatten list genre sambil abaikan NaN
all_genres = pd.Series([genre for sublist in data['genres'] if isinstance(sublist, list) for genre in sublist])

# Ambil 10 genre paling populer
top_genres = all_genres.value_counts().head(10)

print("Top 10 Genres:\n", top_genres)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi rating
plt.figure(figsize=(8,5))
sns.histplot(data['rating'], bins=20, kde=True)
plt.title('Distribusi Rating Film')
plt.xlabel('Rating')
plt.ylabel('Jumlah Film')
plt.show()
```

```python
# One-hot encoding genre
for genre in all_genres.unique():
    data[genre] = data['genres'].apply(lambda x: int(genre in x))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Fitur yang digunakan
features = data[['rating'] + all_genres.unique().tolist()]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# KMeans clustering
kmeans = KMeans(n_clusters=8, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# Lihat hasil clustering
data[['title', 'rating', 'Cluster']].head()
```

```python
# Visualisasi distribusi cluster
plt.figure(figsize=(10,6))
sns.countplot(x='Cluster', data=data, palette='viridis')
plt.title('Distribusi Film per Cluster')
plt.show()
```

```python
def recommend_movies(favorite_movie, n_recommendations=5):
    try:
        # Cari cluster film favorit
        cluster_id = data[data['title'].str.contains(favorite_movie, case=False, regex=False)]['Cluster'].values[0]

        # Ambil film lain di cluster yang sama
        cluster_movies = data[data['Cluster'] == cluster_id]

        # Urutkan berdasarkan rating tertinggi
        recommendations = cluster_movies.sort_values(by='rating', ascending=False)

        # Hapus film favorit dari daftar
        recommendations = recommendations[~recommendations['title'].str.contains(favorite_movie, case=False, regex=False)]

        return recommendations[['title', 'rating']].drop_duplicates().head(n_recommendations)

    except IndexError:
        return "Film tidak ditemukan. Pastikan ejaan judul benar."

# Contoh penggunaan
recommend_movies('Toy Story')
```
